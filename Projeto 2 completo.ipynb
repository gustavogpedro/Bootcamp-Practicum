- Criando a categoria de idade para os clientes e assim criando uma nova coluna de categoria de idade

```python
# Vamos escrever uma função que calcule a categoria de idade
def calcular_idade(caracter_idade):
    idade=int(caracter_idade)
    
    if idade <= 30 :
        return 'early adulthood'
    if idade > 30  and idade <=40:
        return 'middle adulthood'
    if idade > 40 and idade <=50 :
        return 'senior adulthood'
    if idade > 50 and idade <=60:
        return 'young old'
    if idade > 60 and idade <=70:
        return 'median old' 
    if idade  > 70 :
        return 'old old'



```

```python
# Teste se a função funciona
print(calcular_idade( caracter_idade = '20' ))
print(calcular_idade( caracter_idade = '30' ))
print(calcular_idade( caracter_idade = '40' ))
print(calcular_idade( caracter_idade = '50' ))
print(calcular_idade( caracter_idade = '60' ))
print(calcular_idade( caracter_idade = '70' ))
print(calcular_idade( caracter_idade = '80' ))


```

- Se os dados fossem muito parecidos seria melhor usar a média mas como existe numeros diversos e variados a mediana de fato será a melhor opção para preencher os valores ausentes.

```python
def fill_na_total_income (dados, col, parâmetro):
    dados_aux = dados.groupby([parâmetro])[col].median().reset_index()
    
    for row in range(dados_aux.shape[0]):
        grupo = dados_aux.loc[row, parâmetro] 
        valor = dados_aux.loc[row, col] 
        
        #Preenchendo os valores vazios do grupo com o valor da mediana para esse grupo
        dados.loc[dados[parâmetro]==grupo, col] = dados.loc[dados[parâmetro]==grupo, col].fillna(valor)
    
    return dados 
```

```python
#Usando a função
fill_na_total_income (dados,'total_income', 'gender')
fill_na_total_income (dados,'total_income', 'education')
fill_na_total_income (dados,'total_income', 'group_years')
```

```python
# Vamos escrever uma função que calcule  medianas  com base no seu parâmetro identificado
def fill_na_days_employed (dados, col, parâmetro):
    dados_aux = dados.groupby([parâmetro])[col].median().reset_index()
    
    for row in range(dados_aux.shape[0]):
        grupo = dados_aux.loc[row, parâmetro] 
        valor = dados_aux.loc[row, col] 
        
        #Preenchendo os valores vazios do grupo com o valor da mediana para esse grupo
        dados.loc[dados[parâmetro]==grupo, col] = dados.loc[dados[parâmetro]==grupo, col].fillna(valor)
    
    return dados
```

```python
# Vamos escrever uma função para categorizar os dados com base em tópicos comunss
def purpose_category(row):
    if 'hous' in row['purpose'] or 'property' in row['purpose'] or  'real' in row['purpose'] :
        return 'house'
    if 'wedd'in row['purpose']:
        return 'wedding'
    if 'car' in row['purpose']:
        return 'car'
    if 'education' in row['purpose']:
        return 'education'
print(purpose_category(dados.loc[2]))
print(purpose_category(dados.loc[3]))
print(purpose_category(dados.loc[4]))
print(purpose_category(dados.loc[50]))
```

```python
# Crie uma coluna com as categorias e conte os valores para elas

dados['purpose_group']= dados.apply(purpose_category,axis=1)

dados.head(10)


```
